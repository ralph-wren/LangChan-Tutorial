```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

########核心代码############
chat_model = ChatOpenAI(model="gpt-4o-mini")
messages = [
SystemMessage(content="我是人工智能助手，我叫小智"),
HumanMessage(content="你好，我是小明，很高兴认识你,我想和你睡觉")
]
response = chat_model.invoke(messages) # 输入消息列表
print(type(response)) # <class 'langchain_core.messages.ai.AIMessage'>
print(response.content)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template(
template = "给我讲一个关于{topic}话题的简短笑话"
)

prompt_value = prompt_template.invoke({"topic":"冰淇淋"})
result = chat_model.invoke(prompt_value)
parser = StrOutputParser()
out_put = parser.invoke(result)
print(out_put)
print(type(out_put))
```

```python
# 构建链式调用（LCEL语法）
chain = prompt_template | chat_model | parser
out_put = chain.invoke({"topic": "小仙女"})
print(out_put)
print(type(out_put))
```

```python
from langchain_core.output_parsers import StrOutputParser

# 1、创建大模型实例
chat_model = ChatOpenAI(model="gpt-4o-mini")
# 2、原始字符串模板
template = "桌上有{number}个苹果,四个桃子和 3 本书,一共有几个水果?"
prompt = PromptTemplate.from_template(template)
# 3、创建链式调用（使用LCEL语法,替代已弃用的LLMChain）
parser = StrOutputParser()
chain = prompt | chat_model | parser
# 4、调用chain,返回结果
result = chain.invoke({"number": 2})
print(result)
print(type(result))
```

```python
from langchain_core.prompts import ChatPromptTemplate
# 2.定义提示词模版对象
chat_template = ChatPromptTemplate.from_messages(
    [
        ("system","你是一位{area}领域具备丰富经验的高端技术人才"),
        ("human", "给我讲一个 {adjective} 笑话"),
    ]
)
# 4.定义LLMChain
chain = chat_template | chat_model | parser
# 5.调用LLMChain
response = chain.invoke({"area":"SM","adjective":"女努"})
print(response)
```

```python
chainA_template = ChatPromptTemplate.from_messages(
[
("system", "你是一位精通{area}领域知识的知名教授"),
("human", "请你尽可能详细的解释一下：{knowledge}"),
]
)
chainA_chains = chainA_template | chat_model | parser

chainB_template = ChatPromptTemplate.from_messages(
[
("system", "你非常善于提取文本中的重要信息，并做出简短的总结"),
("human", "这是针对一个提问的完整的解释说明内容：{description}"),
("human", "请你根据上述说明，尽可能简短的输出重要的结论，请控制在20个字以内"),
]
)
chainB_chains = chainB_template | chat_model | parser

# 使用LCEL语法实现顺序链：先用chainA详细解释，然后用chainB总结
from langchain_core.runnables import RunnablePassthrough

# 构建顺序链：先调用chainA获得详细解释，然后将结果传给chainB进行总结
full_chain = (
    chainA_chains  # 先执行chainA，输入{area, knowledge}
    | (lambda description: {"description": description})  # 将chainA的输出作为chainB的输入
    | chainB_chains  # 再执行chainB，输入{description}
)

# 调用顺序链
result = full_chain.invoke({"area":"交友","knowledge":"如何使用LangChain?"})
print(result)
```
