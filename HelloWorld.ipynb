```python
# 导入和使用 WebBaseLoader
# 导入和使用 WebBaseLoader
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    web_path="https://blog.xbookcn.net/2015/03/blog-post_61.html",
    bs_kwargs=dict(parse_only=bs4.SoupStrainer(id="UCAP-CONTENT"))
)
docs = loader.load()

# 添加调试信息
print(f"加载的文档数量: {len(docs)}")
if docs:
    print(f"第一个文档的内容长度: {len(docs[0].page_content)}")
    print(f"前500个字符: {docs[0].page_content[:500]}")
else:
    print("警告: 没有加载到任何文档!")

# 对于嵌入模型，这里通过 API调用RAG(检索增强生成)
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")

from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 使用分割器分割文档
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
documents = text_splitter.split_documents(docs)
print(f"分割后的文档数量: {len(documents)}")

# 只有在有文档时才创建向量存储
if documents:
    vector = FAISS.from_documents(documents, embeddings)
    print("向量存储创建成功!")
else:
    print("错误: 没有文档可以创建向量存储，请检查网页加载是否成功")
```
